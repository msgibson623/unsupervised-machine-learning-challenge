```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Import the data
csv = "myopia.csv"

myopia_df = pd.read_csv(csv)
myopia_df
```

```python
# Drop Target
myopia_df2 = myopia_df.drop("MYOPIC", axis=1)
myopia_df2
```

```python
df2scale = StandardScaler().fit_transform(myopia_df2)
print(df2scale.shape)
df2scale[0]
```

```python
pca=PCA(n_components=0.9)
pca_df=pca.fit_transform(df2scale)
print(pca_df.shape)
pca_df[0]
```

```python

pca.explained_variance_ratio_.sum()
```

```python
tsne=TSNE()
tsne_df=tsne.fit_transform(pca_df)
print(tsne_df.shape)
tsne_df[0]
```

```python
x=tsne_df[:,0]
y=tsne_df[:,1]

plt.scatter(x,y)
plt.savefig('scatterFeatureReduction.png')
plt.show()
```

```python
# Identify clusters with k-means
inertia = []
k = list(range(1,11))

for i in k:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(tsne_df)
    inertia.append(kmeans.inertia_)
```

```python
# Create an elbow plot
elbow_df = pd.DataFrame({'k': k, 'inertia': inertia})
elbow_df.plot.line(x='k', y='inertia')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```
